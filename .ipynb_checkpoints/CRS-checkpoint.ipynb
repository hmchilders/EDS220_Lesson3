```python
#import libraries
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt


#update pandas display option
pd.set_option("display.max.columns", None)
```

We want to make a geopnadas geodataframe from our csv data.


```python
#import csv using pandas
power_plants = pd.read_csv('Data/power_plants_epsg4269.csv')
```

```python
power_plants.head(3)
```

From the "documentation" the CRS for this file is EPSG:4269 in lat long coordinates
so we can use this information to create a new geopandas geo data frame from the pandas data frame
To do this we use the 'gpd.points_from_xy' function



```python
power_plants = gpd.GeoDataFrame(power_plants, #data for geodataframe
                               #create/specify the geometry column
                               geometry = gpd.points_from_xy(power_plants.longitude,
                                                            power_plants.latitude),
                               #specifyCRS
                               crs = 'EPSG:4269')

power_plants.head()
```

```python
#print CRS info
print('is geographic:', power_plants.crs.is_geographic)
print('is projected:', power_plants.crs.is_projected)
print('datum:', power_plants.crs.datum)
print('ellipsoid:', power_plants.crs.ellipsoid)
```

```python
states = gpd.read_file('Data/tl_2022_us_states')
```

```python
#simplify colun names
states.columns = states.columns.str.lower()
#check column names
states.head()
```

```python
#What are the datatypes of the columns?
states.dtypes
```

```python
#Update the datatypes of region, division and statefp to be int
states.region = states.region.astype('int64')
states.division = states.division.astype('int64')
states.statefp = states.statefp.astype('int64')
```

```python
states.dtypes
```

```python
#print CRS info
print('is geographic:', states.crs.is_geographic)
print('is projected:', states.crs.is_projected)
print('datum:', states.crs.datum)
print('ellipsoid:', states.crs.ellipsoid)
```

```python
states.plot()
```
